# Mobile Recharge Validation system


```python
Min_Amount = 50
Valid_Plans = [199, 299, 399, 599]

def validate_recharge(amount):
    if amount < Min_Amount:
        return "Rechargr amount must be at least rs.50"

    if amount not in Valid_Plans:
        return f"Invalid plan. Available plans: {Valid_Plans}"

    return "Valid"

while True:
    try:
        amount = int(input("Enter recharge amount : rs. "))
    except ValueError:
        print("Please enter valid numerc amount : ")
        continue

    result = validate_recharge(amount)

    if result == "Valid":
        print("Recharge Successful!")
        break
    else :
        print(result)
        print("Please try again.\n")
```

# Inventory Record Alert System

```python
inventory = {"Rice":10, "Wheat":25, "ToorDal":30, "Moong":8, "Bajra":15 }
def check_inventory(inventory):
    for product , stock in inventory.items():
        if stock < 15 :
            print(f"{product}: Alert (Stock: {stock})")

        else:
            print(f"{product}: Stock is Ok (Stock: {stock}) ")

check_inventory(inventory)
```

# Student Result Processing System

```python
marks = [50, 50 ,70 , 45, 12]
def result_process(marks):
    if not marks:
        return "Marks not provided."
    total = 0
    for mark in marks :
        total += mark

    average =  total / len(marks)

    if average >= 50:
        return f"Average: {average:.2f} --> Pass"
    else :
        return f"Average: {average:.2f} --> Fail"
    
result = result_process(marks)
print(result)
```

# Cab fare Estimator with Retry option

```python
Base_Fare = 50
Per_Km = 12
Peak_Hours = 1.25

def calculate_fare(distance, is_peak):
    fare = Base_Fare + (Per_Km * distance)

    if is_peak:
        fare *= Peak_Hours

    return fare

while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").strip().lower()
        
        if peak_input not in ["yes", "no"]:
            print("Invalid peak hour input. Please enter 'yes' or 'no'.\n")
            continue

        is_peak = peak_input == "yes"

        total_fare = calculate_fare(distance, is_peak)

        print(f"Total Fare: ₹{total_fare:.2f}")

    except ValueError:
        print("Invalid distance input. Please enter a numeric value.\n")
        continue

    retry = input("Do you want to calculate again? (yes/no): ").strip().lower()
    if retry != "yes":
        print("Thank you for using Cab Fare Estimator!")
        break
```

# Employee Attendance Eligiblity Checker

```python
def check_attendance(attendance):
    if not attendance:
        return "No attendance data Provided."
    
    present_count = 0

    for day in attendance:
        if day == "P":
            present_count += 1

    percentage = (present_count / len(attendance)) * 100

    if percentage >= 75:
        return f"Attendance : {percentage: .2f}% --> Eligible"
    else:
        return f"Attendance : {percentage: .2f}% --> Not Eligible"
    
attendance = ["P", "A", "P", "P", "A", "P"]
result = check_attendance(attendance)
print(result)
    
```

# Password strength Checker

```python
password = input("Enter Password:")
def password_strength_checker(Password):
    if len(password) < 8:
        return "Weak Password: Minimum length should be 8 characters."
    
    has_digit =  False
    has_special_character = False

    for char in password:
        if char.isdigit():
            has_digit = True

        if char in "@#$":
            has_special_character = True

    if not has_digit :
        return "Weak Password : Must contain at least one digit"
    
    if not has_special_character:
        return "Weak password: Must contain at least one special character"
    
    return "Strong Password"

result = password_strength_checker(password)
print(result)

```
