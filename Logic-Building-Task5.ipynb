# Social Media - Post Engagement Analyzer

```python
likes = [182, 237, 373 ,233]
def analyze_engagement(likes_list):
    total_likes = 0

    for likes in likes_list:
        total_likes += likes
        
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

analyze_engagement(likes)
    
```

# Agriculture - Rainfall Adequacy Checker

```python
def rainfall_analyzer(rainfall_data, required_level):

    total = sum(rainfall_data)
    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status

rainfall = list(map(float, input("Enter seasonal rainfall values: ").split()))
required_level = float(input("Enter required rainfall level: "))

average, status = rainfall_analyzer(rainfall, required_level)


print(f"Rainfall Status: {status}")
```

# Agriculture - Primium Crop Price Filter

```python
def filter_crops(prices):

    premium_crops = []
    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    return premium_crops

crop_prices = list(map(int, input("Enter Crops Prices : ").split()))

premium = filter_crops(crop_prices)

print(f"Premium Crops: {premium}")
```

# System Monitring - Application Health

```python
def application_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return error_count, status

errors = int(input("Enter Error Count: "))
count, status = application_health(errors)

print (f"Error Count: {count}")
print (f"System Status: {status}")
```

# Banking - Daily Transaction Limit Checker

# E-Learning - Student Attendance Eligib;ity Checker

```python
def check_eligiblity(attendance_list):
    if not attendance_list:
        print("No Attendance Data Provided.")
        return
    present_count = 0

    for day in attendance_list:
        if day == "p":
            present_count += 1

    percentage = (present_count / len(attendance_list)) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print(f"Attendance Percentage: {percentage:.1f}")
    print(f"Exam Eligiblity: {status}")

attendance = input("Enter attendance (P/A): ").split()
check_eligiblity(attendance)
```

# Smart Electricity Bill Analyzer

```python
def calculate_electric_bill(units):
    bill = 0
    if units <= 100:
        bill = units* 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100)* 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status

units_consumed = int(input("Enter Your Consumed Units: "))
bill , usage_status = calculate_electric_bill(units_consumed)

print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{bill} ")
print(f"Usage Status: {usage_status}")
```
