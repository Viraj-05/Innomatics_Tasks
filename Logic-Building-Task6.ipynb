# Smart Parking Lot Management System

```python
parking_logs = ["IN", "IN", "OUT", "IN", "IN", "OUT"]
capacity = 24
def parking_system(logs, capacity):
    count = 0
    peak = 0

    for log in logs:
        if log == "IN":
            count += 1
        elif log == "OUT":
            count -= 1

        if count > peak:
         peak = count

        if count > capacity:
            print("ALERT!! Full Capacity Utilized")

    print(f"Currently Parked Vehicles: {count}")

    if count >= capacity:
        print("Parking is Full !")

    else:
        print("Parking is Available")

    print(f"Peak Parking Usage: {peak} Vehicles")

parking_system(parking_logs, capacity )
```

# Online Food Delivery Time Estimator

```python
distance = float(input("Enter Distance (in km): "))
traffic = input("Enter Traffic Level (Low/Medium/High): ")
weather = input("Whats the Weather (Clear/Cloudy/Rainy): ")

def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5

    if traffic == "Low ":
        traffic_delay = 5
    elif traffic == "Medium":
        traffic_delay = 10
    elif traffic == "High":
        traffic_delay = 15
    else:
        traffic_delay = 0

    if weather == "Clear ":
        weather_delay = 0
    elif weather == "Cloudy":
        weather_delay = 5
    elif weather == "Rainy":
        weather_delay = 10

    else :
        weather_delay = 0

    total_time = base_time + traffic_delay + weather_delay

    return total_time

eta = estimate_delivery_time(distance, traffic, weather)

print(f"Your Order Should Be Delivered In : {eta} minutes")
```

# Movie theater Seat Occupancy Analyzier 

```python
def seat_analyzer (empty_seats, booked_seats):
    total_booked = len(booked_seats)
    occupancy_percentage = (total_booked / empty_seats) * 100

    if occupancy_percentage == 100:
        status = "Housefull"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    elif occupancy_percentage >= 40:
        status = "Moderate Booking"
    else:
        status = "Low Booking"

    if occupancy_percentage >85:
        suggestion = "Consider Opening An New Show"
    else :
        suggestion = "No Additional Shows required"

    return occupancy_percentage , status, suggestion

empty_seats = 200
booked_seats = [1] * 150

occupancy, status, suggestion = seat_analyzer(empty_seats, booked_seats)

print(f"Occupancy: {occupancy: .0f}%")
print(f"Show Status: {status}")
print(f"Recommendation: {suggestion}")

```

# Cloud Server load Classification System

```python
def sever_load_classification(cpu_readings):
    if not cpu_readings:
        print("No CPu data available.")
        return
    
    total = 0
    for reading in cpu_readings:
        total += reading

    average_cpu = total / len(cpu_readings)

    if average_cpu < 50:
        status = "Normal"
    elif average_cpu <= 80:
        status = "Warning High Load"
    else:
        status = "Critical (Power off)"

    return average_cpu, status

cpu_readings = list(map(int,input("Enter CPU Readings : ").split()))

average, status = sever_load_classification(cpu_readings)

print(f"Average CPU Load : {average:.0f} %")
print(f"Server Status: {status}")

```

# Smart Classroom Resource Usage Monitring System

```python
def monitoring_usage(resource_usage):
    overuse_limit = 6
    overused_resources = []

    for resource, hours in resource_usage.items():
        if hours > overuse_limit:
           overused_resources.append(resource)
    
    if overused_resources:
        alert_status = "YES"
    else:
        alert_status = "NO"

    return overused_resources, alert_status

resource_usage = {}

n = int(input("Enter number of resources: "))

for _ in range(n):
    name = input("Enter Resource Name: ")
    hours = float(input("Enter usage hours: "))
    resource_usage[name] = hours

overused, alert = monitoring_usage(resource_usage)

print("Overused Resources:", ",".join(overused) if overused else "None")
print("Usage Alert", alert)

```

# Online Event regestration capacity Controller


```python
capacity = 100
registrations = 122

def regestration_management(capacity, total_regestrations):
    confirmed = 0
    waitlisted = 0

    for _ in range(total_regestrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"
    
    return confirmed, waitlisted, status

confirmed, waitlisted, status = regestration_management(capacity, registrations)

print(f"Confirmed Registrations: {confirmed}")
print(f"Waitlisted Users: {waitlisted}")
print(f"Registration Status: {status}")
```
