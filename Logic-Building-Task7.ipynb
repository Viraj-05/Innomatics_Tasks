# Smart Light Controller

```python
class LightController:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"
    
    def turn_off(self):
        self.status = "OFF"
    
    def display_status(self):
        print(f"{self.name} light is {self.status}")

light_name = input("Enter Lights Name: ")
action = input("ON/OFF : ").upper()

light = LightController(light_name)
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()

light.display_status()
```

# Employee ID card System 

```python
class Employees:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def employee_id_card(self):
        
        print(f"Name        : {self.name}")
        print(f"Employee ID : {self.emp_id}")
        print(f"Department  : {self.department}")
        
employee = Employees("Viraj", "69", "Sales")
employee.employee_id_card()
```

# Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

contact = Contact("Aditya", "98432420047")
contact.display_contact()
```

# Product PriceTag Generator

```python
class Product:
    def __init__(self,name,price):
        self.name =  name
        self.price = price

    def generate_price_tag(self):
        print(f"Product : {self.name}")
        print(f"Price : ₹ {self.price}")

product = Product("Laptop", "64999/-" )
product.generate_price_tag()
```

# Movie Rating Display System

```python
class Rating:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def details(self):
        print(f"Movie Name : {self.name}")
        print(f"Rating : {self.rating}")

movie = Rating("Goodfellas", 9.5)
movie.details()
```

# Delivery Adress Manager

```python
class Delivery :
    def __init__(self, name, address):
        self.name = name
        self.address = address

    def address_manager(self):
        print("Delivery Details")
        print(f"Name : {self.name}")
        print(f"Address : {self.address}")

delivery_address = Delivery("Masoom Sharma", "Haryana Punjab")
delivery_address.address_manager()
```
